# 2. Normalization

> 2.1 Normalization function

2.1.1 ตัวอย่างการเขียนฟังก์ชัน normalization

````python
#nomalization method
def nomalize(X):
  new_X = []
  min_x = min(X)
  max_x = max(X)
  for x in X:
    new_X.append((x - min_x)/(max_x - min_x))

  return new_X
````

2.1.2 ตัวอย่างการเรียกใช้งาน function

````python
#example of using the normalization method
X = [100, 300, 400, 800, 1100]
new_X = nomalize(X)
print(f"input X: {X}")
print(f"Nomalized input (new_X): {new_X}")
````

2.1.3 ให้คำนวณค่า nomalize ของ input X = [200 400 600 800 1600 200 และ 2200]

````python
# ให้เขียน code
X = [200, 400, 600, 800, 1600, 200, 2200]
new_X = nomalize(X)
print(f"input X: {X}")
print(f"Nomalized input (new_X): {new_X}")
````

# 3. การใช้งาน sklearn library สำหรับ k-Nearest Neighbours

> 3.1 k-NN Regression

  3.1.1 ตัวอย่างการสร้างโมเดล และทำนายข้อมูลโดยใช้ 1-NN Regressor (ค่า k = 1)

````python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

#input X of training data
x1 = [0, 1, 2, 5, 6, 7, 8]
X = list(zip(x1))

#output y of traning data
y = [0, 1, 2, 5, 6, 7, 8]

# model
neigh = KNeighborsRegressor(n_neighbors=1)
neigh.fit(X, y)

#predict
x_test = 4.5
y_hat = neigh.predict([[x_test]])

#### Plot data using matplolib ####
#plot x1, x2
plt.scatter(x1, y)
plt.xlabel("input X")
plt.ylabel("output y")

#plot prediction value
plt.scatter([x_test], y_hat, marker='*')
print(f"Prediction output for input x={x_test} is {y_hat}")

plt.show()
````

3.2.5 กำหนดให้ข้อมูลฝึกสอน

𝓣= {[x1 x2], y} = {([0 100],Red), ([0.4 300],0), ([0.6 400],1),([1.4 800],1), ([2.0 1100], 1) }

กำหนดให้ output ของชุดข้อมูลมี 2 คลาสได้แก่ Red = 0, และ Blue = 1

จงหาค่า nomalization ของข้อมูลอินพุต x2





```
*หมายเหตุ example 1 -> ([0 100],Red) ให้เปลี่ยน output y เป็นตัวเลข ทั้งนี้ให้ Red = 0, Blue = 1*

```


จะได้ว่า

𝓣= {[x1 x2], y} = {([0 100],0), ([0.4 300],0), ([0.6 400],1),([1.4 800],1), ([2.0 1100], 1) }



````python
#nomalization method
def normalize(X):
  new_X = []
  min_x = min(X)
  max_x = max(X)
  for x in X:
    new_X.append((x - min_x)/(max_x - min_x))

  return new_X

# 𝓣= {[x1 x2], y} = {([0 100],0), ([0.4 300],0), ([0.6 400],1),([1.4 800],1), ([2.0 1100], 1) }
x1 = [0, 0.4, 0.6, 1.4, 2.0]
x2 = [100, 300, 400, 800, 1100]
classes  = [0, 0, 1, 1, 1]

nor_x2 = normalize(x2)
print (f"x2 = {x2}")
print (f"normalization of x2 = {nor_x2}")
````

3.2.6 จากข้อ 3.2.5 ให้คำนายค่า output y (y_hat) เมื่อ  new input x = [0.35 350] โดยใข้ 1−NN Classifier และ 3-NN Classifier


a. 1−NN Classifier

````python
# ให้เขียน code สำหรับ k = 1
# ตัวอย่างเฉลย
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

def normalize_input (X, input):
  return (input - np.min(X))/(np.max(X) - np.min(X))

data = list(zip(x1, nor_x2))
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(data, classes)

new_x1 = 0.35
new_x2 = 350
nor_new_x2 = normalize_input(x2, new_x2)

new_point = [(new_x1, nor_new_x2)]

prediction = knn.predict(new_point)

plt.scatter(x1 + [new_x1], nor_x2 + [nor_new_x2], c=classes + [prediction[0]])
plt.text(x=new_x1, y=nor_new_x2, s=f"new point, class: {prediction[0]}")
plt.show()
````

b. 3−NN Classifier

````python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

def normalize_input (X, input):
  return (input - np.min(X))/(np.max(X) - np.min(X))

data = list(zip(x1, nor_x2))
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(data, classes)

new_x1 = 0.35
new_x2 = 350
nor_new_x2 = normalize_input(x2, new_x2)

new_point = [(new_x1, nor_new_x2)]

prediction = knn.predict(new_point)

plt.scatter(x1 + [new_x1], nor_x2 + [nor_new_x2], c=classes + [prediction[0]])
plt.text(x=new_x1, y=nor_new_x2, s=f"new point, class: {prediction[0]}")
plt.show()
````
